```python
import pandas as pd
```

```python
x1=pd.read_csv("tfidfconsolidate.csv")
x2=pd.read_csv("w2vconsolidate.csv")
x3=pd.read_csv("fasttextconsolidate.csv")
x4=pd.read_csv("wtw2vconsolidate.csv")
final=x1.filter(['Class'], axis=1)
final['F1_score_tfidf']=x1.filter(['F1_score'],axis=1)
final['F1_score_w2v']=x2.filter(['F1_score'],axis=1)
final['F1_score_fasttext']=x3.filter(['F1_score'],axis=1)
final['F1_score_wtw2v']=x4.filter(['F1_score'],axis=1)
final=final[:-1]
```

```python
l=[]
for i in range(20):
    tmp=[]
    a=final.loc[i,'F1_score_tfidf']
    b=final.loc[i,'F1_score_w2v']
    c=final.loc[i,'F1_score_fasttext']
    d=final.loc[i,'F1_score_wtw2v']
    m=max(a,b,c,d)
    if(m==a):
        tmp.append('TF-IDF')
    if(m==b):
        tmp.append('W2V')
    if(m==c):
        tmp.append('Fasttext')
    if(m==d):
        tmp.append('WtW2V')
    l.append(",".join(tmp))
```

```python
final=final.assign(Winner=l)
```
